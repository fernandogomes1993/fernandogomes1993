```python
class ContaBancaria:
  def __init__(self, id, titular,saldo = 3000) :

    self.id = id
    self.titular = titular
    self.saldo = saldo
    print('>>>'*20)
    print(f'a conta de numero: {self.id} de {self.titular} foi criada com sucesso saldo R$ {self.saldo:,.2f}')
    print('>>>'*20)
    print()

  def depositar(self,valor):
    if valor == 0:
      print()
    else:
      self.saldo += valor
      print(f'o deposito de R$ {valor:,.2f} foi altorizado.... novo saldo de R$ {self.saldo:,.2f}')
      print()

  def sacar(self,valor):
    if valor == 0:
      print()
    elif valor > self.saldo:
      print(f'saldo insuficiente.... para saque saldo atual R$ {self.saldo:,.2f} ')
      print()

    else:
      self.saldo -= valor
      print(f'saque realizado com sucesso.... saldo de R$ {self.saldo:,.2f}')
      print()

  def __str__(self) :
    print('>>>'*20)
    return (f'o numero da conta é {self.id} proprietario {self.titular} com saldo de R${self.saldo:,.2f}')

conta = ContaBancaria(id = input('digite o numero da conta: '),titular= input('nome do titular: '))
conta.depositar(valor = float(input('deseja depositar algum valor: ')))
saque = conta.sacar(valor = float(input('deseja realizar saque qual valor: ')))

print(conta)
```
