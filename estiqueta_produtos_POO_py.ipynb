```python

from rich.panel import Panel
from rich import print

class Produtos:
  def __init__(self,produto,preco):
    self.produto = produto
    self.preco = preco

  def Etiqueta(self):
    conteudo = f'{self.produto.center(30,' ')}'
    conteudo += f'-' * 30
    precof = f'R$ {self.preco:,.2f}'
    print(Panel(f'''{conteudo}
{precof.center(30,"-")}''',title='produto',width=34))

c1 = Produtos('iphone 17 pro max',25_000.85)
c1.Etiqueta()





```
