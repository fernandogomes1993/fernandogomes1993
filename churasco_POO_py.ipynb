```python
# !pip install rich
from rich import print
from rich.panel import Panel

# consumo padrao: 400g por pessoa
# preço R$ 82,40/kg
class Churrasco:
  def __init__(self,titulo,convidados,preco_carne = 82.40 ):
    self.titulo = titulo
    self.convidados = convidados
    self.preco_carne = preco_carne

  def analisar(self):
    carne_necessaria = (400*self.convidados)/1000
    print(Panel(f'''
{'.' * 20} Analizando {'.' * 25}
:white_check_mark: [green]{self.titulo}[/] com [blue]{self.convidados} convidados [/]
:white_check_mark: Cada participante comerá [green]{400/1000}kg [/]de carne ao preço:[green] R$ {self.preco_carne}[/]
:white_check_mark: Recomendo comprar [green]{carne_necessaria:g}kg[/] de carne
:white_check_mark: O custo total será de: [green]R$ {carne_necessaria * self.preco_carne:.2f}[/]
:white_check_mark: Cada convidado pagará [green]R$ {(carne_necessaria * self.preco_carne)/self.convidados:.2f}[/]
        :warning: [red]Caso deseje altera o preço da carne so informa aposa quantidade de convidados[/]''',width=70,title=self.titulo))

c1 = Churrasco('Churrascos dos amigos',15)
c1.analisar()

```
