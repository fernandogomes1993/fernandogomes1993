```python
from rich import print
from rich.panel import Panel


class ControleRemoto:
  canal_min:int = 1
  canal_max: int = 6
  volume_min:int = 1
  volume_max:int = 5

  def __init__(self,canal = 1, volume = 2):
    self.cana_atual:int = canal
    self.volume_atual:int = volume
    self.ligado:bool = False

  def canal_menos(self):
    if self.ligado:
      if self.cana_atual == ControleRemoto.canal_min:
        self.cana_atual = ControleRemoto.canal_max
      else:
        self.cana_atual -= 1
  def canal_mais(self):
    if self.ligado:
      if self.cana_atual == ControleRemoto.canal_max:
        self.cana_atual = ControleRemoto.canal_min
      else:
        self.cana_atual +=1

  def volume_mais(self):
    if self.ligado:
      if self.volume_atual != ControleRemoto.volume_max:
        self.volume_atual +=1
  def volume_menos(self):
    if self.ligado:
      if self.volume_atual == ControleRemoto.volume_menos:
        self.volume_atual = ControleRemoto.volume_max
      else:
        self.volume_atual -=1

  def ligar_desligar(self):
    self.ligado = not self.ligado

  def mostra_tv(self):
    conteudo = ''
    if not self.ligado:
      conteudo += ':prohibited:[red] A TV esta desligada[/]'
    else:
      conteudo = 'CANAL  = '
      for canal in range(ControleRemoto.canal_min,ControleRemoto.canal_max + 1):
        if canal == self.cana_atual:
          conteudo += f'[black on yellow] {canal} [/]'
        else:
          conteudo += f' {canal} '

      conteudo += '\nVOLUME = '
      for volume in range(ControleRemoto.volume_min,ControleRemoto.volume_max + 1):
        if volume <= self.volume_atual:
          conteudo += '[black on cyan] [/]'
        else:
          conteudo += '[black on white] [/]'


      #conteudo += '< CH >   - VOL + '
    tv = Panel(conteudo,title='[ TV ]',width=30)
    print(tv)

c1 = ControleRemoto()
while True:
  c1.mostra_tv()
  comando = str(input('< CH >  - VOL + : '))
  match comando:
    case '@':
      c1.ligar_desligar()
    case  '-':
      c1.volume_menos()
    case '+':
      c1.volume_mais()
    case '<':
      c1.canal_menos()
    case '>':
      c1.canal_mais()
    case  '0':
      break








```
