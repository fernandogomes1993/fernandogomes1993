```python
# instalando a biblioteca de cores e emoji
!pip install rich
from rich import print

class Funcionarios:
  Empresa = 'ABC'
  def __init__(self, nome, setor, cargo) :
    self.nome = nome
    self.setor = setor
    self.cargo = cargo

  def Apresentacao(self):
    return f':handshake: ola, sou [blue] {self.nome} [/] e sou {self.cargo} do setor {self.setor} da empresa {Funcionarios.Empresa}'


c1 = Funcionarios('fernando','administracao','diretor')
# alterando o nome da empresa
#Funcionarios.Empresa = 'KBZ'
print(c1.Apresentacao())
c2 = Funcionarios('joao','TI','programador')
print(c2.Apresentacao())
```
