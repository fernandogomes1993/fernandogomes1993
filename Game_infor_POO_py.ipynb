```python
from rich import print
from rich.panel import Panel

class Gamer:
  def __init__(self,nome,id):
    self.nome = nome
    self.id = id
    self.jogos = []

  def adicionar_Gamer(self,game):
    self.jogos.append(game)

  def ficha(self):
    conteudo = f'Nome Real: [blue]{self.nome}[/]'
    conteudo += '\n -- Jogos Favoritos --'
    for games in self.jogos:
      conteudo += f'\n [blue]:video_game: {games}[/]'

    print(Panel(conteudo,title=f'Jogador < {self.id} >', width=60))


j1 = Gamer('Fernando Gomes','Mathias')
j1.adicionar_Gamer('Fortnite')
j1.adicionar_Gamer('Grand Theft Auto V')
j1.adicionar_Gamer('Call of Duty')
j1.adicionar_Gamer('EA Sports FC 26')

j1.ficha()
```
