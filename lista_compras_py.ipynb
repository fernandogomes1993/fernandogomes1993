```python
lista_compras = []

while True:
  opcao_escolhida = input('''
  Digite [1] Inseir
  Digite [2] Deleta
  Digite [3] Exibir
       ''')


  if opcao_escolhida in '1,2,3':

    if opcao_escolhida == '1':
      adicionar_lista_comprar = input('adicione item a lista: ')
      lista_compras.append(adicionar_lista_comprar)

    elif opcao_escolhida == '2':
      for item in enumerate(lista_compras):
        print(item)
      deletar = int(input('Qual item deseja deletar: '))
      if deletar <= len(lista_compras):
        del lista_compras[deletar]

      else:
        print('>' * 30)
        print('item nao localizado na lista')

    elif opcao_escolhida == '3':
      for item in enumerate(lista_compras):
        print(item)

      break
  else:
    print('>' * 30)
    print('opcao invalida.....')
```
