```python
from time import sleep
from rich import print


class Livro:
  def __init__(self,titulo,paginas):
    self.titulo = titulo
    self.paginas = paginas
    return print(f''':open_book: [red] {self.titulo}[/] ao total sao {self.paginas}''')

  def avanca(self,pagina):
    def avancando_pag(self):
      for contador in range(1,pagina +1):
        print(f'pag{contador}',end='')
        sleep(0.3)

    if pagina <= self.paginas:
      avancando_pag(pagina)
      print(f'\n voçê avancou {pagina} paginas agora esta na pagina {self.paginas - pagina}')
    else:
      pagina = self.paginas
      avancando_pag(pagina)
      print(f':police_car_light: voçê chegou ao final do livro {self.titulo}')


p1 = Livro('10 coisas que aprendi',10)
p1.avanca(5)
p1.avanca(7)


```
