```python
#Vamos criar uma função que padronize códigos de produtos.
# O default será padronizar os códigos para letras minúsculas (dado por 'm'),
# mas se o usuário quiser pode padronizar para maiúscula, dado por ('M').


lista_produtos = [' ABC12 ', 'abc34', 'AbC37']


def padronizando_lista(lista_produtos, padrao):
  for contador, item in enumerate(lista_produtos):
    item = item.replace(' ','')
    item.strip()

    if padrao == 'm':
      item = item.lower()

    elif padrao == 'M':
      item = item.upper()
    lista_produtos[contador] = item
  return lista_produtos

# coloque 'm' para minuscula
# coloque 'M' para maiuscula
padronizando_lista(lista_produtos,padrao = 'm')
```
